```python
import turicreate
sales = turicreate.SFrame('/Users/diegobarbosa/Desktop/Coursera/home_data.sframe/')
```

```python
sales
```

```python
turicreate.show(sales[1:5000]['sqft_living'],sales[1:5000]['price'])
```

```python
train_data, test_data=sales.random_split(0.8,seed=0)
```

```python
test_data.shape
```

```python
train_data.shape
```

```python
#Build regression model
```

```python
sqtf_model=turicreate.linear_regression.create(train_data,target='price',features=['sqft_living'])
```

```python
#Evaluate the simple model
print(test_data['price'].mean())
```

```python
test_data['price'].mean()
```

```python
sqtf_model.evaluate(test_data)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(10,8))
plt.plot(test_data['sqft_living'],test_data['price'],'.', 
         test_data['sqft_living'],sqtf_model.predict(test_data),'-')
plt.xlabel('Square Feet')
plt.ylabel('Price')
plt.grid()
plt.show()
plt.savefig('Linear_reg.pdf')
```

```python
sqtf_model.coefficients
```

```python
#Explore other features
myfeatures=['bedrooms','bathrooms','sqft_living','sqft_lot','floors','zipcode']
sales[myfeatures].show() #selecting only the required features
```

```python
turicreate.show(sales['zipcode'],sales['price']) #Relationship between location and price
```

```python
# Regression model with more features.
newModel=turicreate.linear_regression.create(train_data,target='price',features=myfeatures)
```

```python
print(sqtf_model.evaluate(test_data))
print(newModel.evaluate(test_data))
```
