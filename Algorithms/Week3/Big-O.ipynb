In analyzing algorithms it is common to say that **$f$ grows no faster than $g$** and write $f \preceq g$, if there exists a constant $c$ such that $f(n) \le c \cdot g(n)$ for all $n$.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
n=np.linspace(1,100)
plt.plot(n,7*n**2 +6*n+5)
plt.grid()
plt.show()
```

Observe how $20n$ grows slower than $7n^2 +6n+5$

```python
plt.plot(n, 7*n**2 +6*n+5, 'r' ,label="$7n^2+6n+5$")
plt.plot(n, 20*n, 'g' ,label="$20n$" )
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

### Smaller terms can be ommited :

Consider $7n^2 + 6n +5$ and $7n^2$ for $n \in [1,5]$  

```python
n1=np.linspace(1,5)
plt.plot(n1, 7*n1**2 +6*n1+5, 'r' ,label="$7n^2+6n+5$")
plt.plot(n1, 7*n1**2, 'g' ,label="$7n^2$" )
plt.legend(loc='upper left')
plt.grid()
plt.savefig('PolGrowth1.pdf')
plt.show()
```

But considering $7n^2 + 6n +5$ and $7n^2$ for $n \in [1,100]$, we 
can observe the change:

```python
plt.plot(n, 7*n**2 +6*n+5, 'r' ,label="$7n^2+6n+5$")
plt.plot(n, 7*n**2, 'g' ,label="$7n^2$" )
plt.legend(loc='upper left')
plt.grid()
plt.savefig('PolGrowth2.pdf')
plt.show()
```

Value $n$ at which $n^{0.1}$ becomes larger than $(\log n)^5$.

```python
n = np.linspace(1, 100)
plt.plot(n, n ** .1, label="n^.1")
plt.plot(n, np.log(n) ** 5, label="(log n)^5")
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

```python
n = np.linspace(1, 3)
plt.plot(n, n ** .1, label="$n^{0.1}$")
plt.plot(n, np.log(n) ** 5, label="$(log n)^5$")
plt.legend(loc='upper left')
plt.grid()
plt.show()
```
