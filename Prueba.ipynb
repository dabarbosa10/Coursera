```
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```
with open('train.npy', 'rb') as fin:
    X = np.load(fin)
    
with open('target.npy', 'rb') as fin:
    y = np.load(fin)

plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, s=20)
plt.show()
```

```
def features(X):
  Feat=np.zeros((X.shape[0],6))
  Feat[:,0]=X[:,0]
  Feat[:,1]=X[:,1]
  Feat[:,2]=X[:,0]*X[:,0]
  Feat[:,3]=X[:,1]*X[:,1]
  Feat[:,4]=X[:,0]*X[:,1]
  Feat[:,5]=1
  return Feat
```

```
featu=features(X)
```

```
def logistic(s):
  return 1/(1+np.exp(-s))
```

```
s1=np.arange(-10,10,0.01)
plt.plot(s1,logistic(s1),'r-')
```

```
def prob(X,w):
  X1=features(X)
  return logistic(np.matmul(X1,w))
```
