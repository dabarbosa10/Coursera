```python
#lets look at small cap and large cap companies
returns=mf1[['Lo 10','Hi 10']]
returns.columns=['SmallCap','LargeCap']
returns.head()
```

```python
returns=returns/100
returns.head()
```

```python
returns.index = pd.to_datetime(returns.index, format="%Y%m")
returns.index = returns.index.to_period('M')
returns.head()
```

```python
returns.plot()
```

```python
returns['1960']
```

```python
returns['1960'].plot()
```

```python
wealth_index=1000*(1+returns['LargeCap']).cumprod()
```

```python
wealth_index.plot()
```

```python
wealth_index2=2000*(1+returns['LargeCap']).cumprod()
```

```python
wealth_index.plot()
wealth_index2.plot()
```

```python
previous_peak=wealth_index.cummax()
previous_peak.plot(color='red')
```

```python
drawdown=(wealth_index-previous_peak)/previous_peak
drawdown.plot(color='teal')
```

```python
print(drawdown.min())
print(drawdown.idxmin())
```
